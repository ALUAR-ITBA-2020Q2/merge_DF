```python
import pandas as pd
from matplotlib import pyplot as plt
import sqlite3
```

```python
df_hitm_ps = pd.read_hdf("Datos HITM_PS.hdf", key="df")
```

```python
# strip whitespace from headers
df_hitm_ps.columns = df_hitm_ps.columns.str.strip()
con = sqlite3.connect("database/ALUAR.db")
# drop data into database
df_hitm_ps.to_sql("HITM_PS", con)
con.close()
```
